# Total Return

```python
import yfinance as yf
```

```python
arkk = yf.Ticker('ARKK').history(start='2000-01-01').Close.rename('ARKK')
spy = yf.Ticker('SPY').history(start='2000-01-01').Close.rename('SPY')
brka = yf.Ticker('BRK-A').history(start='2000-01-01').Close.rename('BRKA')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,10))
base = 10_000
ax.plot(arkk/arkk[0]*base, label='ARKK', )
ax.plot(spy/spy[0]*base, label='SPY')
ax.plot(brka/brka[0]*base, label='BRKA')
ax.legend()
```
