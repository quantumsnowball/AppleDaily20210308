# Sharpe Ratio

```python
import yfinance as yf
import numpy as np
```

```python
def sharpe(ticker, start):
    ts = yf.Ticker(ticker).history(start=start).Close.rename(ticker)
    lnchgs = np.log(ts).diff()
    mu = lnchgs.mean() * 252
    sigma = lnchgs.std() * np.sqrt(252)
    sharpe = mu / sigma
    return sharpe
```

```python
print(sharpe('BRK-A', '2015-01-01'))
print(sharpe('SPY', '2015-01-01'))
print(sharpe('ARKK', '2015-01-01'))
```
